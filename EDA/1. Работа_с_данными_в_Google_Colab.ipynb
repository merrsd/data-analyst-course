## Обработка колонки цены — приведение к числовому типу

```python
dataset['Цена']
```

```python
price = '4 000 000 ₽'
```

```python
splited_price = price.split()
print(splited_price)
clean_price = ''.join(splited_price[:-1])
print(clean_price)
```

```python
import re
clean_price = re.sub('\D', '', price)
print(clean_price)
```

```python
def clear_price(price):
  return re.sub('\D', '', price)
```

```python
clear_price(price)
```

 ## Шаблоны анализа

```python
# Шаблон применения какой-либо функции ко всем значениям колонки
import re

def clear_price(price):
  return int(re.sub('\D', '', price))
dataset['Очищенная цена'] = dataset['Цена'].apply(clear_price)
```

```python
# Шаблон сортировки по значению
sorted_dataset = dataset.sort_values('Очищенная цена')
sorted_dataset
```
